# The pitcher's example

```python
"""Gaussian mean-field variational parameterization.

Defines functions to initialize variational parameters, sample epsilon noise, compute zeta samples, and compute entropy of the variational posterior. Used in variational inference with a Gaussian approximation to the posterior.
"""
# Gaussian mean-field variational parameterization
def vp_init(size):
    """Initialize variational parameters for a Gaussian distribution.
    
    Args:
      size: The size of the Gaussian distribution. 
    
    Returns:
      mu: The mean of the Gaussian distribution.
      log_s: The log standard deviation of the Gaussian distribution.
    """
    mu = np.random.randn(size)
    log_s = np.random.randn(size)  # omega
    return jnp.array([mu, log_s])

"""Gaussian mean-field variational parameterization.

Defines functions to sample epsilon noise, compute zeta samples, 
and compute entropy of the variational posterior. Used in variational 
inference with a Gaussian approximation to the posterior."""
def eps_rvs(size, n=1):
    """Sample epsilon noise for variational inference.
    
    Args:
      size: The size of the Gaussian distribution.
      n: The number of samples.
      
    Returns:
      epsilons: The sampled epsilon noise, with shape (n, size).
    """
    return np.random.randn(n, size)

def zeta(vp, eps):
    """Compute zeta samples for variational inference.
    
    Given variational parameters and epsilon noise, compute samples 
    from the variational posterior (zeta) using the reparameterization
    trick.
    
    Args:
      vp: Variational parameters (mu, log_s)
      eps: Epsilon noise
      
    Returns:
      zeta: Samples from the variational posterior  
    """
    mu = vp[0]
    s = jnp.exp(vp[1])
    return mu + s * eps   

def entropy_q(vp):
    """
    Compute the entropy of a Gaussian variational posterior.
    
    Args:
      vp: Variational parameters (mu, log_s) for the Gaussian posterior.
      
    Returns:
      Entropy of the Gaussian posterior.
    """

    K = len(vp[0])
    return K / 2. * (1. + np.log(2 * np.pi)) + jnp.sum(vp[1]) 
```

```python
# Transformation from bounded to real
def sigmoid(x):
    return 1. / (1. + jnp.exp(-x))
    
def T(z, a, b):
    zeta = jsc.special.logit((z - a) / (b - a))
    return zeta

def T_inv(zeta, a, b):
    z = a + (b - a) * sigmoid(zeta)
    return z
    
def det_jac_T_inv(zeta, a, b):
    return (b - a) * sigmoid(zeta) * (1 - sigmoid(zeta))
```

```python
# Prior
def log_prior_v(v):
    return jnp.log(1. / 10)

def log_prior_alpha(alpha):
    return jnp.log(2. / np.pi)
```

```python
# Likelihood
def _trajectory(v, alpha, dt=0.1, G=9.81):
    v_x = v * jnp.cos(alpha)  # x velocity m/s
    v_y = v * jnp.sin(alpha)  # y velocity m/s
    
    x = 0.0  # initial position
    y = 1.5  # initial height

    while y > 0:  # simulate until ball hits floor
        v_y += dt * -G  # acceleration due to gravity
        x += dt * v_x
        y += dt * v_y
        
    return x, y

def log_likelihood(x_obs, v, alpha, sigma=0.5):
    mu, _ = _trajectory(v, alpha)
    return -0.5 * ((x_obs - mu) / sigma) ** 2 - jnp.log(sigma * (2. * np.pi) ** 0.5)
```

```python
def elbo(vp, x_obs, eps):
    """Computes the ELBO for the variational posterior over the parameters,
    given the observed data x_obs and noise eps.
    
    The ELBO is the objective function we aim to maximize when fitting
    the variational posterior. It is a lower bound on the marginal
    likelihood of the data.
    
    Args:
      vp: Variational parameters
      x_obs: Observed data
      eps: Noise
    
    Returns:
      ELBO scalar
    """    _zeta = zeta(vp, eps)
    out = 0.0
    
    # log p(v)
    v = T_inv(_zeta[0], 0., 10.)
    out += log_prior_v(v)
    out += jnp.log(det_jac_T_inv(_zeta[0], 0., 10.))
    
    # log p(alpha)
    alpha = T_inv(_zeta[1], 0., np.pi / 2)
    out += log_prior_alpha(alpha)
    out += jnp.log(det_jac_T_inv(_zeta[1], 0., np.pi / 2.))
    
    # log p(x|v, alpha)
    out += log_likelihood(x_obs, v, alpha)
    
    # entropy q
    out += entropy_q(vp)
    
    return jnp.sum(out) 

grad_elbo = jax.grad(elbo)
```

```python
v_true = 7.5
alpha_true = np.pi / 4
mu, _ = _trajectory(v_true, alpha_true)
x_obs = np.random.randn(1) * 0.5 + mu
x_obs = x_obs[0]

vp = vp_init(2)
batchsize = 25
lr = 0.02
n_epochs = 50

wandb.init(project="nb08-advi")
wandb.config.update({
    "v_true": v_true, "alpha_true": alpha_true, "x_obs": x_obs,
    "lr": lr, "batchsize": batchsize
})
```

```python
test = eps_rvs(2, 100)
g = grad_elbo(vp, x_obs, test[0])

for epoch in tqdm(range(n_epochs)):
    # Log
    elbo_test = np.mean([elbo(vp, x_obs, eps) for eps in test])
    wandb.log({"ELBO": elbo_test})
    
    # Update            
    g = jnp.zeros_like(g)
    
    for eps in eps_rvs(2, batchsize):
        g += grad_elbo(vp, x_obs, eps)
    
    vp += lr * g / batchsize
```

```python
_zeta = zeta(vp, eps_rvs(2, 10000))
v = T_inv(_zeta[:, 0], 0., 10.)
alpha = T_inv(_zeta[:, 1], 0., np.pi / 2.)
```

```python
import corner
thetas = np.array([v, alpha]).T
labels = ["v", "alpha"]
fig = corner.corner(thetas, labels=labels, truths=[v_true, alpha_true])
```

```python
# p(x | x_obs)
xs = []
    
for v, alpha in thetas[:5000]:
    mu, _ = _trajectory(v, alpha)   
    x = np.random.randn(1) * 0.5 + mu
    xs.append(x[0])

plt.hist(xs, density=True, bins=30, color="orange")
plt.scatter([x_obs], [0.])
plt.show()
```
